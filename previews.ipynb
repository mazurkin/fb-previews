# Частотный метод

```python
# Параметры размытия
dim_size = (24, 24)
dim_blur = (5, 5)
```

```python
# Функция для фильтрации компонентной плоскости картинки
def filter_fft(img, name):
    c_data = np.array(img, dtype=np.uint8)

    # FFT
    c_fft = np.fft.fft2(c_data)

    # разделяем гармоники на амплитуду и фазу
    c_abs = np.abs(c_fft)
    c_ang = np.angle(c_fft)

    # вычисляем порог ниже которого находится большая часть гармоник (по амплитуде)
    abs_limit = np.percentile(c_abs, 96)

    # сбрасываем все гармоники ниже этого порога
    m = (c_abs <= abs_limit)
    c_abs[m] = 1
    c_ang[m] = 0

    # квантизируем амплитуду с переводом в логарифмическую шкалу
    c_abs = np.where(c_abs > 1, np.log(c_abs), 0)
    c_abs_steps = 32
    c_abs_step = c_abs.max() / (c_abs_steps - 1)
    c_abs_data = np.round(c_abs / c_abs_step).astype(np.int8)

    # квантизируем фазу (-PI..+PI)
    c_ang_steps = 32
    c_ang_step = 2 * np.pi / (c_ang_steps - 1)
    c_ang_data = np.round(c_ang / c_ang_step).astype(np.int8)

    # сохраняем
    c_abs_data.tofile('data/' + name + '-abs.dat')
    c_ang_data.tofile('data/' + name + '-ang.dat')

    # количество ненулевых коэффициентов
    print("non zero cells={0}".format(np.count_nonzero(c_abs_data)))

    # восстановление амплитуды 
    c_abs = c_abs_step * c_abs_data.astype(np.float)
    c_abs = np.where(c_abs > 0, np.exp(c_abs), 0)

    # восстановление угла
    c_ang = c_ang_step * c_ang_data.astype(np.float)

    # возращаем к комплексному виду
    c_fft_rst = np.multiply(c_abs, np.exp(1j * c_ang))

    # IFFT
    c_ifft = np.fft.ifft2(c_fft_rst)
    c_ifft_abs = np.abs(c_ifft)
    c_ifft_int = c_ifft_abs.astype(np.uint8)
    
    # Сглаживаем плоскость отдельно
    c_ifft_blur = cv2.GaussianBlur(c_ifft_int, dim_blur, 0)

    return c_ifft_blur
```

```python
# Исходная картинка
im_src = Image.open('i/fb-original.jpg')
im_prw = Image.open('i/fb-preview.jpg')

print('source size is {0}'.format(im_src.size))
```

```python
# Уменьшаем
im_dim = np.array(im_src, dtype=np.uint8)
im_dim = cv2.resize(im_dim, dim_size, interpolation=cv2.INTER_AREA)
```

```python
# Сглаживаем и переводим в HLS
im_dat = cv2.GaussianBlur(im_dim, dim_blur, 0)
im_dat = cv2.cvtColor(im_dat, cv2.COLOR_BGR2HLS)
```

```python
# Фильтруем плоскости по отдельности (считаем количество ненулевых коэффициентов)
s1_s, s2_s, s3_s = cv2.split(im_dat)
s1_t, s2_t, s3_t = filter_fft(s1_s, 's1'), filter_fft(s2_s, 's2'), filter_fft(s3_s, 's3')
```

```python
# Собираем плоскости обратно из HLS
merged = cv2.merge((s1_t, s2_t, s3_t))
merged = cv2.cvtColor(merged, cv2.COLOR_HLS2BGR)
```

```python
# Увеличиваем до размера первоначальной картинки и сглаживаем
merged_inc = cv2.resize(merged, im_src.size,interpolation=cv2.INTER_LANCZOS4)
merged_inc = cv2.GaussianBlur(merged_inc, (99, 99), 0)
```

```python
# Сравниваем исходную и восстановленную

plt.subplot(1, 3, 1)
plt.imshow(im_src)
plt.title('Original FB')
plt.xticks([])
plt.yticks([])

plt.subplot(1, 3, 2)
plt.imshow(im_prw)
plt.title('Preview FB')
plt.xticks([])
plt.yticks([])

plt.subplot(1, 3, 3)
plt.imshow(merged_inc)
plt.title('Merged')
plt.xticks([])
plt.yticks([])

plt.show()
```

# Простой метод

```python
# Очень сильно уменьшаем картинку
im_ico = np.array(im_src, dtype=np.uint8)
im_ico = cv2.resize(im_ico, (4, 4), interpolation=cv2.INTER_AREA)
```

```python
# Оставляем 4 бита из восьми
im_ico_4 = np.bitwise_and(im_ico, 240)
```

```python
# Увеличиваем до размера первоначальной картинки и сглаживаем
im_ico_rst = cv2.resize(im_ico_4, im_src.size,interpolation=cv2.INTER_LANCZOS4)
```

```python
# Сравниваем исходную и восстановленную

plt.subplot(1, 3, 1)
plt.imshow(im_src)
plt.title('Original FB')
plt.xticks([])
plt.yticks([])

plt.subplot(1, 3, 2)
plt.imshow(im_prw)
plt.title('Preview FB')
plt.xticks([])
plt.yticks([])

plt.subplot(1, 3, 3)
plt.imshow(im_ico_rst)
plt.title('Restored')
plt.xticks([])
plt.yticks([])

plt.show()
```
